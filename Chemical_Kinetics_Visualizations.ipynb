```python
import seaborn as sns 
import numpy
import pandas as pd
print(numpy.__version__, pd.__version__)
```

```python
data = pd.read_csv('OCAT_Survey_Responses_Form_responses.csv')
```

```python
data
```

```python
data.columns = [col.strip().replace(':','').replace(' ','_').replace('"','').lower() for col in data.columns]

data
```

```python
# Convert necessary columns to numeric
numeric_columns = ['age', 'experience_with_current_organization_(in_years)', 
                   'community_and_collaboration', 'innovation_and_agility', 
                   'competitive_drive', 'structure_and_control', 'task_performance']

data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
# Converts invalid values to NaN 

# Check data summary
print(data.describe())
```

```python
# Correlation matrix
correlation_matrix = data[['community_and_collaboration', 'innovation_and_agility',
                           'competitive_drive', 'structure_and_control', 'task_performance']].corr()

# Display correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

# Pairwise scatter plots
sns.pairplot(data, vars=data[['community_and_collaboration', 'innovation_and_agility', 
                         'competitive_drive', 'structure_and_control', 'task_performance']])
plt.show()


```

```python
# Task performance by designation
plt.figure(figsize=(10, 6))

#OUTLIERS
sns.boxplot(x='designation', y='task_performance', data=data)
plt.xticks(rotation=90)
plt.title('Task Performance by Designation')
plt.show()


plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(x='experience_with_current_organization_(in_years)', y='task_performance', data=data, hue='designation')
# Move the legend outside the plot
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Designation')
plt.title('Task Performance vs. Experience')
plt.xlabel('Experience with Current Organization (in years)')
plt.ylabel('Task Performance')
plt.show()
```

```python
# Boxplot for task performance by education level
plt.figure(figsize=(10, 6))
sns.boxplot(x='education', y='task_performance', data=data)
plt.xticks(rotation=45)
plt.title('Task Performance Across Education Levels')
plt.show()
```

```python
# Identify outliers in task performance
plt.figure(figsize=(10, 6))
sns.boxplot(y='task_performance', data=data)
plt.title('Outliers in Task Performance')
plt.show()
```

```python
plt.hist(data['task_performance'], bins=50)
plt.xlabel('Task Performance')
plt.ylabel('Frequency')
plt.title('Histogram of Task Performance')
plt.show()
```

```python
plt.boxplot([data['community_and_collaboration'], data['innovation_and_agility'], data['competitive_drive'], data['structure_and_control']])
plt.xticks([1, 2, 3, 4], ['Community & Collaboration', 'Innovation & Agility', 'Competitive Drive', 'Structure & Control'], rotation=45, ha='right')
plt.ylabel('Rating')
plt.title('Box Plot of Cultural Dimensions')
plt.show()
```

```python
# columns 6 to 11 represent cultural dimensions
culture_cols = data.iloc[:, 5:11]
culture_averages = culture_cols.mean(axis=0)

plt.figure(figsize=(10, 6))
culture_averages.plot(kind='bar', color=['skyblue', 'lightgreen', 'gold', 'coral', 'purple'])
plt.title("Average Scores for Organizational Culture Dimensions")
plt.xlabel("Dimension")
plt.ylabel("Average Score")
plt.xticks(rotation=90)

# Adjust margins to prevent overlapping labels
plt.subplots_adjust(bottom=0.25) 

plt.show()
```
